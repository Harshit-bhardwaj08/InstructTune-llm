```python
# Superdense Coding Protocol

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

def superdense(bits):
    qc = QuantumCircuit(2, 2)

    # Bell pair
    qc.h(0)
    qc.cx(0, 1)

    # Encoding
    if bits == "01":
        qc.x(0)
    elif bits == "10":
        qc.z(0)
    elif bits == "11":
        qc.x(0)
        qc.z(0)

    # Decoding
    qc.cx(0, 1)
    qc.h(0)

    qc.measure([0,1], [0,1])

    # Simulator
    sim = AerSimulator()
    compiled = transpile(qc, sim)
    result = sim.run(compiled, shots=1024).result()

    return result.get_counts()

# Test
for b in ["00","01","10","11"]:
    print(b, superdense(b))
```

```python
# Quantum Teleportation
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(3, 3)

# Step 1: Prepare |ψ⟩ (example: superposition)
qc.h(0)

# Step 2: Create Bell pair
qc.h(1)
qc.cx(1, 2)

# Step 3: Bell measurement
qc.cx(0, 1)
qc.h(0)

qc.measure(0, 0)
qc.measure(1, 1)

# Step 4: Conditional correction (quantum equivalent)
qc.cx(1, 2)
qc.cz(0, 2)

# Step 5: Measure final qubit
qc.measure(2, 2)

# Run on simulator
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

print(result.get_counts())
```

```python
# Quantum Coin Game
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

sim = AerSimulator()

# Case 1: B does nothing
qc1 = QuantumCircuit(1, 1)

qc1.h(0)
# B does nothing
qc1.h(0)

qc1.measure(0, 0)

compiled1 = transpile(qc1, sim)
result1 = sim.run(compiled1, shots=1024).result()
print("No flip:", result1.get_counts())


# Case 2: B flips
qc2 = QuantumCircuit(1, 1)

qc2.h(0)
qc2.x(0)   # B flips
qc2.h(0)

qc2.measure(0, 0)

compiled2 = transpile(qc2, sim)
result2 = sim.run(compiled2, shots=1024).result()
print("Flip:", result2.get_counts())
```

```python
# Phase Kickback

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2, 2)

# Example input: |10>
qc.x(0)

# Apply H on both qubits
qc.h([0,1])

# Normal CNOT
qc.cx(0,1)

# Apply H again
qc.h([0,1])

# Measure
qc.measure([0,1], [0,1])

sim = AerSimulator()
result = sim.run(transpile(qc, sim), shots=1024).result()

print("Reverse CNOT:", result.get_counts())
```

```python
# Phase Kickback

from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2)

# Step 1: Prepare superposition
qc.h(1)

# Step 2: Entangle
qc.cx(0,1)

# Step 3: Apply T gate on target
qc.t(1)

# Step 4: Undo entanglement
qc.cx(0,1)

# Get statevector
state = Statevector.from_instruction(qc)

print("Statevector with phase kickback:")
print(state)
```

```python
# Deutsch Algorithm

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

def deutsch(oracle_type):
    qc = QuantumCircuit(2, 1)

    # Step 1: Initialize
    qc.x(1)
    qc.h([0,1])

    # Step 2: Oracle
    if oracle_type == "balanced":
        qc.cx(0,1)   # f(x) = x
    elif oracle_type == "constant_1":
        qc.x(1)      # f(x) = 1
    # constant_0 → do nothing

    # Step 3: Interference
    qc.h(0)

    # Measure
    qc.measure(0,0)

    sim = AerSimulator()
    result = sim.run(transpile(qc, sim), shots=1024).result()
    return result.get_counts()

# Test
print("Constant (0):", deutsch("constant_0"))
print("Constant (1):", deutsch("constant_1"))
print("Balanced:", deutsch("balanced"))
```

```python
# Deutsch-Jozsa Algorithm

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

def deutsch_jozsa(n, oracle_type):
    qc = QuantumCircuit(n+1, n)

    # Step 1: Initialize
    qc.x(n)
    qc.h(range(n+1))

    # Step 2: Oracle
    if oracle_type == "balanced":
        for i in range(n):
            qc.cx(i, n)   # balanced function
    elif oracle_type == "constant_1":
        qc.x(n)
    # constant_0 → do nothing

    # Step 3: Interference
    qc.h(range(n))

    # Step 4: Measure
    qc.measure(range(n), range(n))

    sim = AerSimulator()
    result = sim.run(transpile(qc, sim), shots=1024).result()
    return result.get_counts()

# Test with n=3
print("Constant:", deutsch_jozsa(3, "constant_0"))
print("Balanced:", deutsch_jozsa(3, "balanced"))
```

```python
# Quantum Key Distribution

from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile
import random

n = 10

alice_bits = [random.randint(0,1) for _ in range(n)]
alice_bases = [random.choice(['Z','X']) for _ in range(n)]

qc = QuantumCircuit(n, n)

# Encode
for i in range(n):
    if alice_bits[i] == 1:
        qc.x(i)
    if alice_bases[i] == 'X':
        qc.h(i)

# Bob bases
bob_bases = [random.choice(['Z','X']) for _ in range(n)]

for i in range(n):
    if bob_bases[i] == 'X':
        qc.h(i)

qc.measure(range(n), range(n))

sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1).result()

counts = result.get_counts()
measured = list(counts.keys())[0][::-1]

key = []
for i in range(n):
    if alice_bases[i] == bob_bases[i]:
        key.append(int(measured[i]))

print("Shared Key:", key)
```

```python
# BB84

import random
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile

n = 12

alice_bits = [random.randint(0,1) for _ in range(n)]
alice_bases = [random.choice(['Z','X']) for _ in range(n)]

eve_bases = [random.choice(['Z','X']) for _ in range(n)]
bob_bases = [random.choice(['Z','X']) for _ in range(n)]

qc = QuantumCircuit(n, n)

# Alice
for i in range(n):
    if alice_bits[i] == 1:
        qc.x(i)
    if alice_bases[i] == 'X':
        qc.h(i)

# Eve
for i in range(n):
    if eve_bases[i] == 'X':
        qc.h(i)

# Bob
for i in range(n):
    if bob_bases[i] == 'X':
        qc.h(i)

qc.measure(range(n), range(n))

sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1).result()

counts = result.get_counts()
measured = list(counts.keys())[0][::-1]

errors = 0
key = []

for i in range(n):
    if alice_bases[i] == bob_bases[i]:
        key.append(int(measured[i]))
        if int(measured[i]) != alice_bits[i]:
            errors += 1

print("Key:", key)
print("Errors:", errors)
```

```python
# Grover’s Search

from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile

qc = QuantumCircuit(2, 2)

# Superposition
qc.h([0,1])

# Oracle
qc.cz(0,1)

# Diffusion
qc.h([0,1])
qc.x([0,1])
qc.h(1)
qc.cx(0,1)
qc.h(1)
qc.x([0,1])
qc.h([0,1])

qc.measure([0,1], [0,1])

sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1000).result()

counts = result.get_counts()
print(counts)
```
